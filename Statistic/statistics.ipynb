## STATISTICS

## measure of spread(dispersion)

```python
import numpy as np
import statistics as stats
li=np.random.randint(0,100,10)
print(li)
print("mean:",np.mean(li))
print("median:",np.median(li))
print("mode:",stats.mode(li))
```

## INTERQUARTILE RANGE(IQR)

```python
#step 1:
import numpy as np
import scipy.stats as stats
#step 2:
data=[12,15,14,10,18,20,22,25,14,17]
#step 3:
range_value=max(data)-min(data)
print(f"range:{range_value}")
```

```python
#step 4 variance
variance_value=np.var(data,ddof=0) #population variance
sample_variance_value=np.var(data,ddof=1) #sample variance
print(f"population variance:{variance_value}")
print(f"sample_variance:{sample_variance_value}")
```

```python
#step 5: standared deviation
std_dev_value=np.std(data,ddof=0) #population standared deviation
sample_std_dev_value=np.std(data,ddof=1) #sample standared deviation

print(f"population standared deviation:{std_dev_value}")
print(f"sample standared deviation:{sample_std_dev_value}")
```

```python
# step 6:
Q1=np.percentile(data,25) #first quartile(25th percentile)
Q3=np.percentile(data,75) #third quartile(75th percentile)
iqr_value=Q3-Q1
print(f"interquartile range(IQR):{iqr_value}")
```
